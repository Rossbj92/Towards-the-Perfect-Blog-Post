# Description

This notebook houses all of the code used to determine final model selection and performance on test data. 

````python
#Data manipulation
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.options.mode.chained_assignment = None 

#Visualization
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import matplotlib.patches as mpatches
import seaborn
%matplotlib inline
plt.style.use('seaborn')
#Fonts for plots
header_font = fm.FontProperties(fname='../Fonts/Oswald-VariableFont_wght.ttf', size = 14)
text_font = fm.FontProperties(fname='../Fonts/SairaCondensed-Regular.ttf', size = 14)
legend_font = fm.FontProperties(fname='../Fonts/SairaCondensed-Regular.ttf', size = 20)

#Modeling
import scipy.stats as stats
import statsmodels.api as sm
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet, RidgeCV, PoissonRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, r2_score, mean_squared_error
from sklearn.model_selection import train_test_split, KFold, cross_val_predict, cross_validate, GridSearchCV, cross_val_score
````

````python
data = pd.read_pickle('../Data/medium_2020_processed.pkl', compression = 'bz2')
data.head(1)
````

````python
X = data.drop(['Unnamed: 0', 'date', 'title', 'post', 'claps', 'tokens', 'log_claps'], axis = 1)
y = data['claps']
````

````python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=13)
````

The curent features are currently on various types of scales. To immediately begin comparing coefficients, I use scikit-learn's ```StandardScaler()``` to put them on equivalent scales (i.e., mean of 0 and variance of 1). Next, I use Statsmodels for a preliminary linear regression model, as this will give a clean layout of overall model statistics and feature importances.

````python
scaler = StandardScaler()

#Formatting back to Pandas to df to re-add in column names instead of X1, X2, X3, etc.
X_train = pd.DataFrame(scaler.fit_transform(X_train))
X_train.columns = X.columns

X_test = scaler.transform(X_test)
````

````python
X_train_const = sm.add_constant(X_train)
#Reset y_train index so indices match with X_train_const
claps_model_const = sm.OLS(np.log1p(y_train.reset_index(drop = True)), X_train_const)

results_const = claps_model_const.fit()

results_const.summary()
````

<br>
<br>
The majority of the features are significant, but p-values are dangerous as a sole gauge. This is especially so when we use large samples, since standard errors will naturally decrease with sample size and make magically low p-values easier to come by. <br><br>

A better metric here is r-squared. At .09, this is telling us that despite our copious significant features, the model itself is capturing approximately 9% of the total variance in posts' claps. In other words, we're not doing a great job at predicting claps! <br>

The nature of these data is such that linear regression may not be the most ideal model. That is, post claps is a positive integer target variable in which the distribution may be more suited to a generalized linear model. One of these, Poisson regression, which is traditionally used on count data, could more accurately model the right-skewed distribution that we observed in the previous notebook. For this model, we'll need to use the raw claps rather than log transformed since it needs integer values.

````python
claps_model_pois_const = sm.Poisson(y_train.reset_index(drop = True), X_train_const)

results_pois_const = claps_model_pois_const.fit()

results_pois_const.summary()
````

Looking at the r-squared values from both models, it initially appears that a Poisson model is a better fit to the data. However, modeling in sklearn using kfold cross validation showed that this did not hold up. A more extensive model selection methodology is displayed later, but I used these initial findings to guide me towards sticking with selections of linear models. 

````python
kf = KFold(n_splits=5, shuffle=True, random_state = 71)
print('Poisson r2 mean: {:.3f}'.format(np.mean(cross_val_score(PoissonRegressor(alpha = 0, max_iter = 1000), 
                                                               X_train, 
                                                               y_train, 
                                                               cv = kf, 
                                                               scoring = ("r2")))))
print('Linear Regression r2 mean: {:.3f}'.format(np.mean(cross_val_score(LinearRegression(), 
                                                                         X_train, 
                                                                         np.log1p(y_train), 
                                                                         cv = kf, 
                                                                         scoring = ("r2")))))
````

<br>
<br>
Returning to linear regression, perhaps it could be that a handful of features are useful, but I am adding many that are not and unknowingly introducing noise into the model. Below, I loop through and fit a linear regression model to each feature individually.

````python
for i in range(len(X_train_const.columns)):
    
    claps_model_const_loop = sm.OLS(np.log1p(y_train.reset_index(drop = True)), X_train_const.iloc[:,i])

    results_const_loop = claps_model_const_loop.fit()

    print('Feature: {} \n R2: {:.3f}'.format(X_train_const.columns[i], results_const_loop.rsquared))
````

<br>
<br>
Unfortunately nothing was standing out. I tested many subsets and combinations of features, and the best model contained all of them. I thus moved forward without dropping any.
<br>
There was no question that assumptions were going to be bent here. To get a better gauge of how the simple linear regression model was behaving, I utilized the residuals and Q-Q plots below. 

````python
fig, ax = plt.subplots(1,2, figsize=(20,8))

ax[0].scatter(results_const.fittedvalues, np.log1p(y_train.reset_index(drop = True)) - results_const.fittedvalues)
ax[0].set_xlabel('Predicted')
ax[0].set_ylabel('Residuals')
ax[0].set_title('Predicted vs residuals')

stats.probplot(np.log1p(y_train.reset_index(drop = True)) - results_const.fittedvalues, dist = 'norm', plot = plt)
ax[1].set_title("Normal Q-Q plot")

fig.suptitle('OLS Diagnostic Plots', fontsize = 20)
plt.savefig('diagnostic_plots.png');
````

The residual plot isn't a picture-perfect textbook example. There's a hard cutoff that we see due to the impossibility of negative values, and this inherently makes the residuals non-normal. Regardless, the residuals are fairly randomly distributed. There does still remain some heteroscedasticity with the variance in the bigger predictions larger than the rest, so we'll need to keep this in mind. The Q-Q plot is not perfect either, but we do not drastic departures from an expected normal distribution.  

Following evaluation of the assumptions for a linear model, I test several below. Additionally, I decided to test a Random Forest model since these are more robust to both outliers and non-normality. 

````python
poly = PolynomialFeatures()
X_train_poly = poly.fit_transform(X_train)
````

````python
model_choices = {'Linear Regression':LinearRegression(), 
                 'Polynomial Regression':LinearRegression(),
                 'Ridge':Ridge(random_state = 0), 
                 'Lasso':Lasso(random_state = 1),
                 'ElasticNet':ElasticNet(random_state = 1), 
                 'RandomForest': RandomForestRegressor(random_state = 2)}

modelcomp_df = pd.DataFrame({'Model':[], 
                             'R2':[], 
                             'MAE':[]}).set_index('Model')

for name,instance in model_choices.items(): 
    if name != 'Polynomial Regression':
        cv = cross_validate(instance, 
                       X_train,
                       np.log1p(y_train),
                       cv = 5, 
                       scoring = ('r2', 'neg_mean_absolute_error', 'neg_mean_squared_error'))
    else:
        cv = cross_validate(instance, 
                       X_train_poly,
                       np.log1p(y_train),
                       cv = 5, 
                       scoring = ('r2', 'neg_mean_absolute_error', 'neg_mean_squared_error')) 
        
    modelcomp_df.loc[name, 'R2'] = np.mean(cv['test_r2'])
    modelcomp_df.loc[name, 'MAE'] = -np.mean(cv['test_neg_mean_absolute_error'])
    modelcomp_df.loc[name, 'MSE'] = -np.mean(cv['test_neg_mean_squared_error'])
````

````python
modelcomp_df.sort_values('R2', ascending = False)
````

Based on these results, I'm going to move forward with tuning a random forest and ridge regression model. Before predicting the test set, I'll tune the hyperparameters. While the random forest did perform the best, its results will be slightly less interpretable than a ridge regression. The emphasis for this current project is a coherent story, so as long as the differences in performance are not striking, ridge would be preferable in this context.

## Model tuning

### Random forest tuning

````python
#Code adapted from 'Hands on Machine Learning' (p. 76)
param_grid = [
    {'n_estimators': [3, 10, 30], 'max_features': [2, 4, 6, 8]},
    {'bootstrap': [False], 'n_estimators': [3, 10], 'max_features': [2, 3, 4]},
]

forest_reg = RandomForestRegressor(random_state = 3)

rf_grid_search = GridSearchCV(forest_reg, param_grid, cv=5, scoring='r2',
                           return_train_score=True)

rf_grid_search.fit(X_train, np.log1p(y_train))

rf_cv = cross_val_score(rf_grid_search.best_estimator_, X_train, np.log1p(y_train), cv = 5, scoring = 'r2')

print('Mean R2: {:.3f}'.format(np.mean(rf_cv)))
````

### Ridge tuning

````python
alphavec = 10**np.linspace(-2,2,200)

ridge_model = RidgeCV(alphas = alphavec, cv=10)

ridge_model.fit(X_train, np.log1p(y_train))

ridge_cv = cross_val_score(ridge_model, X_train, np.log1p(y_train), cv = 5, scoring = 'r2')

print('Mean R2: {:.3f}'.format(np.mean(ridge_cv)))
````

## Predicting test set

After extensive EDA, feature engineering, model selection, and iterations of all the aforementioned, we can finally move onto predicting the test data!

````python
test_preds = ridge_model.predict(X_test)
print('Test R2: {:.3f}'.format(r2_score(np.log1p(y_test), test_preds)))
print('Test MAE: {:.3f}'.format(mean_absolute_error(y_test, np.expm1(test_preds))))
````

That was a bit anticlimatic - albeit, not unexpectedly.
